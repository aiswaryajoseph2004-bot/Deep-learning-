# Reshaping & Manipulating Tensors

### 1.   tf.reshape() - Changes tensor shap
2.   tf.transpose() - Transposes a matrix
3.  tf.squeeze() - Removes dimensions of size 1
4.  tf.expand_dims() - Adds a new dimension

```python
import tensorflow as tf
tensor = tf.constant([1, 2, 3, 4, 5, 6])
reshaped_tensor = tf.reshape(tensor, [2, 3])
print(reshaped_tensor)
```

```python
import tensorflow as tf
matrix = tf.constant([[1, 2, 3], [4, 5, 6]])
transposed = tf.transpose(matrix)
print(transposed)
```

```python
import tensorflow as tf
tensor = tf.constant([[[1], [2], [3]]])
squeezed = tf.squeeze(tensor)
print(squeezed)
```

```python
import tensorflow as tf
tensor = tf.constant([1, 2, 3])
expanded = tf.expand_dims(tensor, axis=1)
print(expanded)
```

# Aggregation Functions (Summarizing Data)

### 1. tf.reduce_sum() - Computes sum of all elements
2. tf.reduce_mean() - Computes mean (average)
3. tf.reduce_max() & tf.reduce_min() - Finds max/min


```python
import tensorflow as tf
tensor=tf.constant(([1,2],[3,4]))
sum=tf.reduce_sum(tensor)
print(sum)
```

```python
mean_result=tf.reduce_mean(tensor)
print(mean_result)
```

```python
mean_result=tf.reduce_min(tensor)
print(mean_result)
```

# feedforward

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
x_train=np.array(([1],[2],[3],[4],[5]),dtype=float)
y_train=np.array([[3],[5],[7],[9],[11]],dtype=float)
```

```python
model=Sequential([
    Dense(10,activation='relu',input_shape=(1,)),
    Dense(1)
])
```

```python
model.compile(optimizer=Adam(learning_rate=0.01),loss='mean_squared_error')
```

```python
model.fit(x_train,y_train,epochs=100,verbose=1)
```

```python
x_test=np.array([[6],[7]],dtype=float)
y_pred=model.predict(x_test)

print("predictions for 6 and 7:",y_pred)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
x_test=np.array([[6],[7]],dtype=float)
y_test=np.array([[13],[15]],dtype=float)
y_pred=model.predict(x_test)
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
print(f"predictions:{y_pred.flatten()}")
print(f"Mean Squared Error:{mse}")
print(f"Mean Absolute Error:{mae}")
```

## classification

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

x,y=make_classification(n_samples=1000,n_features=20,n_classes=2,random_state=42)

scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)

x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.3,random_state=42)
```

```python
model = Sequential()
model.add(Dense(64, input_dim=20, activation='relu'))  # Hidden Layer 1
model.add(Dense(32, activation='relu'))               # Hidden Layer 2
model.add(Dense(1, activation='sigmoid'))             # Output Layer

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=10, batch_size=32)
```

```python
probs=model.predict(np.expand_dims(x_test[0],axis=0))
pred_classes=np.round(probs).astype(int)

print('probabilities:\n',probs.flatten())
print()
print('predicted classes(by rounding):',pred_classes.flatten())
```

```python
loss,accuracy=model.evaluate(x_test,y_test)
print(f"Model Accuracy:{accuracy * 100:.2f}%")
```
